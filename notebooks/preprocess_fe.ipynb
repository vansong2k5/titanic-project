```python
data_df = pd.read_csv('../data/train.csv', index_col='PassengerId')
```

```python
list_columns = list(data_df.columns)
# list_columns.remove('Name')
# list_columns.remove('Age')
for i in list_columns:
    print(data_df[i].value_counts())
    print('\n')
```

```python
list_columns = list(data_df.Name.values)
missing_title = []

for i in list_columns:
    # Kiểm tra xem có bất kỳ title nào trong chuỗi i không
    if not any(title in i for title in ['Mr.', 'Mrs.', 'Miss.', 'Master.']):
        missing_title.append(i)

if len(missing_title) == 0:
    print("Tất cả đều có dấu hiệu Mr, Mrs, Miss...")
else:
    print(f"Có {len(missing_title)} dòng thiếu danh xưng:")
    print(missing_title)
```

```python
# 'SibSp', 'Parch' = FamilySize
# Name = Tittle ( Mr, Mrs, Master) 
features = ['Pclass','Sex', 'Embarked']
def convert_cat(df, features):
    for feature in features:
        df[feature] = df[feature].astype('category')
convert_cat(data_df,features)
```
