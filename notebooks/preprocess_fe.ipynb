```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_df = pd.read_csv('../data/train.csv', index_col='PassengerId')
```

#### Create new value
- Create FamilySize = SibSp + Parch + 1
- Create IsAlone = 1 if FamilySize == 1 else 0

```python
data_df['FamilySize'] = data_df['SibSp'] + data_df['Parch'] + 1
data_df['IsAlone'] = (data_df['FamilySize']==1).astype(int)
data_df.head()
```

#### Handle Tittle from Name by regex

```python
# Chuyển kiểu dữ liệu sang Category
data_df.Survived = data_df.Survived.astype("category")
```

```python
data_df.info()
```

### Quy trình đúng nên là:

- Phân loại (Numeric vs Categorical).

- Phân loại con (Nominal vs Ordinal, Discrete vs Continuous).

- Xử lý dữ liệu thiếu (Imputation).

- Encoding & Scaling.

```python
# 'SibSp', 'Parch' = FamilySize
# Name = Tittle ( Mr, Mrs, Master) 
features = ['Pclass','Sex', 'Embarked']
def convert_cat(df, features):
    for feature in features:
        df[feature] = df[feature].astype('category')
convert_cat(data_df,features)
```
